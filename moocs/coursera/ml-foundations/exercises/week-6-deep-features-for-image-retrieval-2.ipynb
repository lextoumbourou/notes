```python
import graphlab
```

```python
# Load the CIFAR-10 dataset
```

```python
image_train = graphlab.SFrame('../courseware/image_train_data/')
```

```python
image_train.head()
```

```python
# Train nearest-neighbours model for retrieving images with deep features
```

```python
knn_model = graphlab.nearest_neighbors.create(
    image_train, features=['deep_features'], label='id')
```

```python
# Use image retrieval model with deep features to find similiar images
```

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(
        query_result['reference_label'], 'id')
```

```python
cat_neighbours = get_images_from_ids(knn_model.query(cat))
```

```python
cat_neighbours['image'].show()
```

```python
car = image_train[8:9]
car['image'].show()
```

```python
car_neighbours = get_images_from_ids(knn_model.query(car))
```

```python
car_neighbours['image'].show()
```

```python
# Find and show nearest neighbour images
```

```python
show_nearest_neighbours = lambda img: image_train.filter_by(
    knn_model.query(img)['reference_label'], 'id')['image'].show()
```
