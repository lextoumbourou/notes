```python
import graphlab
```

```python
sales = graphlab.SFrame('home_data.gl/')
```

```python
sales
```

```python
sales.show(view='Scatter Plot', x='sqft_living', y='price')
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
sales.show(view='Scatter Plot', x='sqft_living', y='price')
```

```python
# Create simple regression model of sqft_living to price
```

```python
train_data,test_data = sales.random_split(.8, seed=1234)
```

```python
train_data.show(view='Scatter Plot', x='sqft_living', y='price')
```

```python
test_data.show(view='Scatter Plot', x='sqft_living', y='price')
```

```python
# Build the regression model
```

```python
sqft_model = graphlab.linear_regression.create(train_data, target='price', features=['sqft_living'])
```

```python
# Evaluate the simple model
```

```python
print test_data['price'].mean()
```

```python
sqft_model.evaluate(test_data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_data['sqft_living'],test_data['price'],'.',test_data['sqft_living'],sqft_model.predict(test_data),'-')
```

```python
sqft_model.get('coefficients')
```
