```python
import numpy as np
```

```python
def entropy(dist):
    return -np.sum(np.array(dist) * np.log2(dist))
```

An example of a higher entropy distribution:

```python
dist = [1/10.]*10
dist
```

```python
entropy(dist)
```

An example of a lower entropy distribution:

```python
dist = [99/100] + [1/100.]
dist
```

```python
entropy(dist)
```
